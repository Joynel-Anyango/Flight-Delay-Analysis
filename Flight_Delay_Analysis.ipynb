## Dataset Cleaning

```python
#Checking missing Values
FLIGHT.isnull().sum()
```

```python
#Checking for duplicates
FLIGHT.duplicated().sum()
```

```python
#Dealing with date column
#Creating Month_year column from year and month columns
FLIGHT['month_year'] = pd.to_datetime(FLIGHT[['Year', 'Month']].assign(day=1)).dt.strftime('%b-%Y')
#Creating full_date column from year and month columns
FLIGHT['full_date'] = pd.to_datetime(FLIGHT[['Year', 'Month']].assign(day=1))
FLIGHT
```

```python
FLIGHT.to_excel('Cleaned_Airline_Data.xlsx', index=False)
```

## Exploratory Data Analysis

### 1. Flight Arrivals

### A. Total Flight Arrivals

```python
#Total Arrivals
FLIGHT['Total Flights Arrival'].sum()
```

### B. Previous and Current Month Flight Arrivals

```python
# sum of Flight arrivals grouped by month_year
monthly_summary = FLIGHT.groupby('full_date')['Total Flights Arrival'].sum().reset_index()

# previous month's arrivals
monthly_summary['prev_month_arrivals'] = monthly_summary['Total Flights Arrival'].shift(1)

# Viewing the result
print(monthly_summary)
```

```python
# Visualization of Monthly Flight Arrivals vs Previous Month
plt.figure(figsize=(12, 6))
plt.plot(monthly_summary['full_date'], monthly_summary['Total Flights Arrival'], marker='o', label='Total Arrivals')
plt.plot(monthly_summary['full_date'], monthly_summary['prev_month_arrivals'], marker='x', linestyle='--', label='Previous Month Arrivals')

plt.title('Monthly Flight Arrivals vs Previous Month')
plt.xlabel('Month-Year')
plt.ylabel('Number of Arrivals')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

### C. Previous and Current Year Flight Arrivals

```python
#  sum of Flight arrivals grouped by month_year
Yearly_summary = FLIGHT.groupby('full_date')['Total Flights Arrival'].sum().reset_index()

# previous Year's arrivals
Yearly_summary['prev_year_arrivals'] = Yearly_summary['Total Flights Arrival'].shift(12)

# Viewing the result
print(Yearly_summary)
```

```python
# Visualization of Yearly Flight Arrivals vs Previous Year
plt.figure(figsize=(12, 6))
plt.plot(Yearly_summary['full_date'], Yearly_summary['Total Flights Arrival'], marker='o', label='Total Arrivals')
plt.plot(Yearly_summary['full_date'], Yearly_summary['prev_year_arrivals'], marker='x', linestyle='--', label='Previous Year Arrivals')

plt.title('Yearly Flight Arrivals vs Previous Year')
plt.xlabel('Month-Year')
plt.ylabel('Number of Arrivals')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.show()
```

### D. Total Arrivals per Airport

```python
#Total Arrivals by Airport
FLIGHT.groupby('Airport Name')['Total Flights Arrival'].sum().sort_values(ascending=False)
```

### E. Total Arrivals per Carrier

```python
#Total Arrivals by Carrier
FLIGHT.groupby('Carrier Name')['Total Flights Arrival'].sum().sort_values(ascending=False).plot(kind='bar',color='green',title='Total Arrivals by Carrier')
```
